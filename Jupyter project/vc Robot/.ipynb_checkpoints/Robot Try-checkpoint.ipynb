Ваша задача — на основе исторических данных максимально точно предсказать коэффициент работоспособности: target<br>
Исторические данные для обучения и предсказания даны в хронологическом порядке.<br>
Решением будет файл формата .csv с двумя колонками: year и target,<br>
где предсказаны значения работоспособности для каждого года из тестового набора данных.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image

import warnings
warnings.filterwarnings('ignore')
```

```python
df_test=pd.read_csv('test_data1.csv', sep=';', index_col='year')
df_train = pd.read_csv('train_data.csv', sep=',', index_col='year')
df_train.head()
```

```python
df_train.columns
```

```python
# Объем данных
df_train.shape
```

```python
# Статистика
df_train.describe()
```

```python
data_corr=df_train.corr()

plt.figure(figsize = (10,6))
sns.heatmap(data_corr)
pass
```

```python
X_train = df_train.loc[:,df_train.columns != 'target']
X_train = X_train.loc[:,X_train.columns != 'gamma_ray']
X_train_k = df_train.gamma_ray

X_test = df_test.loc[:,df_test.columns != 'target']
X_test = X_test.loc[:,X_test.columns != 'gamma_ray']
X_test_k = df_test.gamma_ray

X_test.head()
```

```python
Y_train = df_train[['target']]

Y_train.head()
```

```python
# Обработка вещественных данных

from sklearn.preprocessing import StandardScaler

ss=StandardScaler()

#X_train=ss.fit_transform(X_train)
#X_test=ss.transform(X_test)

print('Минимум и максимум обучающего набора: ', X_train.min(), X_train.max())
print('Минимум и максимум тестового набора: ', X_test.min(), X_test.max())
```

```python
# Обработка категориальных данных
X_train_k = X_train_k.map({'high': 2, 'moderate': 1, 'low': 0})
X_test_k = X_test_k.map({'high': 2, 'moderate': 1, 'low': 0})
```

```python
# Объединяем вещественные и категориальные данные
X_train_all= X_train.join(X_train_k)
X_test_all= X_test.join(X_test_k)
```

```python
#Check on NAN and INF
#test
#np.any(np.isfinite(X_test_all))
#np.all(np.isnan(X_test_all))
#np.where(pd.isnull(X_test_all))
#cleaning
X_test_all.gamma_ray.fillna(X_test_all.gamma_ray.mean(), inplace=True)
#X_test_all.info()
X_test_all.isnull().sum()
```

```python
plt.figure(figsize=(14,7))
plt.plot(X_train_all.gamma_ray[30:], 'b')
plt.plot(X_test_all.gamma_ray[30:], 'r')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
pass
```

```python
X_train_all.shape
```

### Базовое решение

```python
from sklearn.metrics import mean_absolute_error

Y_mean=Y_train.rolling(window=30).mean()

print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[30:], Y_mean[30:]), 2))
print('Среднее значение : ', np.round(Y_train[30:].mean(), 2))
```

```python
plt.figure(figsize=(14,7))
plt.plot(Y_train[30:], 'b')
plt.plot(Y_mean[30:], 'r')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

```python
plt.figure(figsize=(14,7))
plt.plot(Y_train[30:] - Y_mean[30:], 'g')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

### Линейная регрессия

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train_all, Y_train)
```

```python
Pred_lr_train=lr.predict(X_train_all)
Pred_lr_test=lr.predict(X_test_all) #have large value??
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[30:], Pred_lr_train[30:]), 2))
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[2030:3000], Pred_lr_test[30:]), 2))
```

```python
plt.figure(figsize=(14,7))
x=range(len(Y_train))
plt.plot(x[30:], Y_train[30:], 'b')
#plt.plot(x[30:], Y_mean[30:], 'r')
plt.plot(x[30:], Pred_lr_train[30:], 'g')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

```python
plt.figure(figsize=(14,7))
plt.plot(Y_train[30:] - Pred_lr_train[30:], 'g')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

### Регуляризация

```python
# Эластичная регрессия (L1 и L2 - регуляризация)

from sklearn.linear_model  import ElasticNet

lr_elastic=ElasticNet(alpha=0.05, l1_ratio=0.005)
lr_elastic.fit(X_train_all, Y_train)
```

```python
Pred_lr_elastic_train=lr_elastic.predict(X_train_all)
Pred_lr_elastic_test=lr_elastic.predict(X_test_all)
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[30:], Pred_lr_elastic_train[30:]), 2))
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[2030:3000], Pred_lr_test[30:]), 2))
```

```python
plt.figure(figsize=(14,7))
plt.plot(Y_train[30:], 'b')
plt.plot(Pred_lr_elastic_train[30:], 'g')
plt.plot(Pred_lr_elastic_test[30:], 'r')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

```python
# Гребневая регрессия (L2 - регуляризация)

from sklearn.linear_model  import Ridge

lr_ridge=Ridge(alpha=25)
lr_ridge.fit(X_train_all, Y_train)
```

```python
Pred_lr_ridge_train=lr_ridge.predict(X_train_all)
Pred_lr_ridge_test=lr_ridge.predict(X_test_all)
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[30:], Pred_lr_ridge_train[30:]), 2))
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[2030:3000], Pred_lr_test[30:]), 2))
```

```python
plt.figure(figsize=(14,7))
plt.plot(Y_train[30:], 'b')
plt.plot(Y_train[30:] - Pred_lr_ridge_train[30:] +17, 'g')  #+ coefficient!!!!!!!!
plt.plot(Pred_lr_ridge_test[30:], 'b')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

```python
# Эластичная регрессия (L1 и L2 - регуляризация) от данных линейной

from sklearn.linear_model  import ElasticNet

lr_elastic2=ElasticNet(alpha=0.05, l1_ratio=0.005)
lr_elastic2.fit(X_train_all, Pred_lr_train)
```

```python
Pred_lr_elastic_train2=lr_elastic2.predict(X_train_all)
Pred_lr_elastic_test2=lr_elastic2.predict(X_test_all)
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[30:], Pred_lr_elastic_train2[30:]), 2))
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[2030:3000], Pred_lr_elastic_test2[30:]), 2))
```

```python
plt.figure(figsize=(14,7))
x=range(len(Y_train))
plt.plot(x[30:], Y_train[30:], 'b')
plt.plot(x[30:], Y_train[30:] - Pred_lr_train[30:],'r')
#plt.plot(x[30:], Pred_lr_train[30:], 'g')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

```python
# Лассо регрессия (L1 - регуляризация)

from sklearn.linear_model  import Lasso

lr_lasso=Lasso(alpha=0.15)
lr_lasso.fit(X_train_all, Y_train)
```

```python
Pred_lr_lasso_train=lr_lasso.predict(X_train_all)
Pred_lr_lasso_test=lr_lasso.predict(X_test_all)
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[30:], Pred_lr_lasso_train[30:]), 2))
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[2030:3000], Pred_lr_lasso_test[30:]), 2))
```

```python
plt.figure(figsize=(14,7))
x=range(len(Y_train))
plt.plot(x[30:], Y_train[30:], 'b')
#plt.plot(x[30:], Y_train[30:] - Pred_lr_lasso_train[30:],'r')
plt.plot(x[30:], Pred_lr_lasso_train[30:], 'g')
plt.xlabel('delta_year')
plt.ylabel('target, eq')
plt.grid()
```

```python
# Лассо регрессия (L1 - регуляризация) от данных линейной 

from sklearn.linear_model  import Lasso

lr_lasso2=Lasso(alpha=0.15)
lr_lasso2.fit(X_train_all, Pred_lr_train)
```

```python
Pred_lr_lasso_train2=lr_lasso.predict(X_train_all)
Pred_lr_lasso_test2=lr_lasso.predict(X_test_all)
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[30:], Pred_lr_lasso_train2[30:]), 2))
print('Среднее абсолютное отклонение: ', np.round(mean_absolute_error(Y_train[2030:3000], Pred_lr_lasso_test2[30:]), 2))
```
